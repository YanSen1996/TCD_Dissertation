```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img  = cv2.imread('1.jpg', 0)

img_1 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                              cv2.THRESH_BINARY, 11, 2)

img_2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                              cv2.THRESH_BINARY, 11, 2)

ret, img = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)

titles = ['GUASSIAN','MEAN','Simper']
imgs =[img_1, img_2, img]

fig = plt.figure(figsize=(15, 15))

for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(imgs[i],'gray')
    plt.title(titles[i])

plt.show()
```

```python
img_can = cv2.Canny(img_1, 50, 150)
cont, hier = cv2.findContours(np.array(img_can), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
len(cont)
```

```python
img_can = cv2.Canny(img_2, 50, 150)
cont, hier = cv2.findContours(np.array(img_can), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
len(cont)
```
